```python
print('Broadcasted element-wise equality comparison:')
d = torch.tensor([[1., 2.], [3., 4.]])
e = torch.ones(1, 2) #many comparison ops support broadcasting
print(torch.eq(d, e)) #returns a tensor of type bool
```

```python
print('Reduction ops:')
print(torch.max(d)) #returns a single element tensor ie max
print(torch.max(d).item()) #extracts the value of the returned tensor
print(torch.mean(d))
print(torch.std(d))
print(torch.prod(d))
print(torch.unique(torch.tensor([1, 2, 1, 2, 1, 2]))) #filter unique elements
```

```python
#vector and linear algebra operations
v1 = torch.tensor([1., 0., 0.]) #x unit vector
v2 = torch.tensor([0., 1., 0.]) #y unit vector
m1 = torch.rand(2, 2) #random matrix
m2 = torch.tensor([[3., 0.], [0., 3.]]) #three times identity matrix
print('\nVectors and Matrices:')
print(torch.linalg.cross(v2, v1)) #negative of z unit vector(v1 x v2 == -v2 x v1)
print(m1)
m3 = torch.linalg.matmul(m1, m2)
print(m3)
print(torch.linalg.svd(m3))
```

##Broadcasting
Each tensor must have at least one dimension - no empty tensors.

Comparing the dimension sizes of the two tensors, going from last to first:

1. Each dimension must be equal, or

2. One of the dimensions must be of size 1, or

3. The dimension does not exist in one of the tensors

```python
a = torch.ones(4, 3, 2)
print(a.shape)

b = a * torch.rand(  3, 2)
print(b.shape)

c = a * torch.rand(  3, 1)
print(c.shape)

d = a * torch.rand( 1, 2) #all work because dimensions match up or 1
print(d.shape)
```

```python
a = torch.ones(2, 4)
print(a)

b = torch.ones(2, 1)* 2
print(b)

c = a * b
print(c)
print(c.shape)
```

##Altering Tensors in place
1. Math Operators - use an '_'

```python
a = torch.tensor([0, math.pi / 4, math.pi / 2, 3 * math.pi / 2, 3 * math.pi / 4 ])
print('a:')
print(a)
print(torch.sin(a)) #new operation in memory
print(a) #a unchanged
```

```python
b = torch.tensor([0, math.pi / 4, math.pi / 2, 3 * math.pi / 2, 3 * math.pi / 4 ])
print('b:')
print(b)
print(torch.sin_(b)) #underscore maxxing
print(b) #b changed
```

```python
a = torch.ones(2, 2)
b = torch.ones(2, 2)

print('Before')
print(a)
print(b)
print('\nAfter adding:')
print(a.add_(b))
print(a)
print(b)
print('\nAfter multiplying:')
print(b.mul_(b))
print(b)
```
