```python
import torch
```

```python
x = torch.empty(3, 4)
print(type(x))
print(x)
```

```python
zeros = torch.zeros(2, 3)
print(zeros)

ones = torch.ones(2, 3)
print(ones)

torch.manual_seed(42)
random = torch.rand(2,3)
print(random)
```

```python
x = torch.empty(2, 3, 3)
print (x.shape)
print(x)

empty_like_x = torch.empty_like(x)
print(empty_like_x.shape)
print(empty_like_x)

zeros_like_x = torch.zeros_like(x)
print(zeros_like_x.shape)
print(zeros_like_x)

ones_like_x = torch.ones_like(x)
print(ones_like_x.shape)
print(ones_like_x)

rand_like_x = torch.rand_like(x)
print(rand_like_x.shape)
print(rand_like_x)
```

```python
some_constants = torch.tensor([[3.14, 2.56], [6.52, 4.53]])
print(some_constants)

some_integers = torch.tensor((1, 2, 3, 4, 5, 6, 7, 8))
print(some_integers)

more_integers = torch.tensor(((1, 2, 3), [4, 5, 6]))
print(more_integers)
```

```python
a = torch.ones((2, 3), dtype = torch.int16)
print(a)

b = torch.rand((2, 3), dtype=torch.float64) * 20
print(b)

c = b.to(torch.int32)
print(c)
```
