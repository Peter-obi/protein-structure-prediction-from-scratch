##Broadcasting
Each tensor must have at least one dimension - no empty tensors.

Comparing the dimension sizes of the two tensors, going from last to first:

1. Each dimension must be equal, or

2. One of the dimensions must be of size 1, or

3. The dimension does not exist in one of the tensors

```python
a = torch.ones(4, 3, 2)
print(a.shape)

b = a * torch.rand(  3, 2)
print(b.shape)

c = a * torch.rand(  3, 1)
print(c.shape)

d = a * torch.rand( 1, 2) #all work because dimensions match up or 1
print(d.shape)
```

##Altering Tensors in place
1. Math Operators - use an '_'. Note these in-place arithmetic functions are methods on the torch.Tensor object, not attached to the torch module like many other functions (e.g., torch.sin()). As you can see from a.add_(b), the calling tensor is the one that gets changed in place.
2. Use the out argument

```python
a = torch.tensor([0, math.pi / 4, math.pi / 2, 3 * math.pi / 2, 3 * math.pi / 4 ])
print('a:')
print(a)
print(torch.sin(a)) #new operation in memory
print(a) #a unchanged
```

```python
b = torch.tensor([0, math.pi / 4, math.pi / 2, 3 * math.pi / 2, 3 * math.pi / 4 ])
print('b:')
print(b)
print(torch.sin_(b)) #underscore maxxing
print(b) #b changed
```

```python
a = torch.ones(2, 2)
b = torch.ones(2, 2)

print('Before')
print(a)
print(b)
print('\nAfter adding:')
print(a.add_(b))
print(a)
print(b)
print('\nAfter multiplying:')
print(b.mul_(b))
print(b)
```

```python
a = torch.rand(2, 2)
b = torch.rand(2, 2)
c = torch.rand(2, 2)
old_id = id(c)
print(old_id)

print(c)
d = torch.matmul(a, b, out = c)
print(c)

assert c is d
assert id(c) == old_id

torch.rand(2, 2, out  = c) #works for creation
print(c)
assert id(c) == old_id
```

##Copying tensors
 Assigning a tensor to a variable makes the variable a label of the tensor, and does not copy it. If you want a separate copy of the data to work on, use clone() method.  

```python
a = torch.ones(2, 2)
b = a

a[0][1] = 561 # change a
print(b) #b is altered
```

```python
a = torch.ones(2, 2)
b = a.clone()

assert b is not a #different objects in memory
print(torch.eq(a, b)) #same contents

a[0][1] = 561
print(a) #a changes
print(b) #b remains same
```

```python
#to prevent clone from tracking gradients, use 'detach()'

a = torch.rand(2, 2, requires_grad=True) #turn on autograd
print(a)

b = a.clone()
print(b)

c = a.detach().clone()
print(c)

print(a)
```

##Manipulating tensor shapes
- Use unsqueeze and squeeze to change dimensions

```python
a = torch.rand(3, 226, 226)
b = a.unsqueeze(0) #add a new zeroth dimension

print(a.shape)
print(b.shape)
```

```python
#squeeze and unsequeeze take advantage of the fact that any dimension of extent 1 does not change the number of elements in a tensor
a = torch.rand(1, 20)
print(a.shape)
print(a)

b = a.squeeze(0)
print(b.shape) #now one dimension, one square bracket lost
print(b)

c = torch.rand(2, 2)
print(c.shape)

d = c.squeeze(0) #cant squeeze any dimension more than one
print(d.shape)
```

```python
a = torch.ones(4, 3, 2)
b = torch.rand( 3)
c = b.unsqueeze(1)
print(c.shape)
print(a * c)
```

```python
#has in place method, using '_'
batch_me = torch.rand(3, 226, 226)
print(batch_me.shape)
batch_me.unsqueeze_(0)
print(batch_me.shape)
```

```python
output_3d = torch.rand(6, 20, 20)
print(output_3d.shape)

input1d = output_3d.reshape(6 * 20 * 20) #move from 3D to 1D. remmeber reshape shares same underlying memory, so clone if you need to remain same
print(input1d.shape)

input1d = output_3d.reshape(-1) #does same job
print(input1d.shape)

print(torch.reshape(output_3d, (6 * 20 * 20,)).shape)
```

##Numpy bridge

```python
import numpy as np

numpy_array = np.ones((2, 3))
print(numpy_array)

pytorch_tensor = torch.from_numpy(numpy_array)
print(pytorch_tensor)
```

```python
pytorch_rand = torch.rand(2, 3)
print(pytorch_rand)

numpy_rand = pytorch_rand.numpy()
print(numpy_rand)
```

```python
#change to one affects the other
numpy_array[1, 1] = 13
print(pytorch_tensor)

pytorch_rand[1, 1] = 17
print(numpy_rand)
```
